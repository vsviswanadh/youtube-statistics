```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb 
```

```python
df=pd.read_csv("Global YouTube Statistics.csv",encoding="ISO-8859-1")
```

```python
pd.set_option("display.float_format","{:.2f}".format)
```

```python
category_mean=df.groupby("category")["subscribers"].mean().sort_values(ascending=False)
category_mean
```

```python
category_mean.head().plot(kind="bar")
plt.title("mean subscribers by category")
plt.ylabel("mean subscribers")
plt.show()
```

```python
channel_type_count=df["channel_type"].value_counts()
channel_type_count
```

```python
channel_type_count.head().plot(kind="pie",autopct="%1.1f%%")
plt.show()
```

```python
plt.scatter(df["subscribers"],df["video views"])
plt.show()
```

```python
df.columns
```

```python
total_subscribers=df.groupby("Country")["subscribers"].sum().sort_values(ascending=False)
total_subscribers
```

```python
total_subscribers.head().plot(kind="bar",color="red",edgecolor="black")
plt.title("total subscribers by country")
plt.xlabel("country")
plt.ylabel("subscribers")
plt.xticks(rotation=0)
plt.show()
```

```python
avg_subscribers=df.groupby("created_year")["subscribers"].mean()
avg_subscribers
```

```python
df.groupby(["category","Country"])["subscribers"].sum().unstack().plot(kind="bar",stacked=True)
plt.legend(title="country",bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
```
